```python
#Importing pandas for data maniputaltion
import pandas as pd
df=pd.read_csv("merged_movies2.csv")
```

```python
#Preprocessing
df['year'] = pd.to_datetime(df['release_date']).dt.year
df['decade']=(df['year']//10)*10
df=df.dropna(subset='decade')
```

```python
#weighted mean rating calculation
c = df['vote_average'].mean()
m= df['vote_count'].mean()
def weightedmeanrating(x):
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m) * r) + (m/(m+v) * c)
df['weighted_rating'] = df.apply(weightedmeanrating, axis=1)
```

```python
#Top movies per decade calculation
def topmovies_perdecade(df):
    topgrossing_perdecade = df.sort_values(['decade', 'box_office'], ascending=[True, False]).groupby('decade').head(10)

    toprated_perdecade = df[df['vote_count'] >= 100] .sort_values(['decade', 'weighted_rating'], ascending=[True, False]).groupby('decade').head(10)

    return topgrossing_perdecade, toprated_perdecade
topgrossing, toprated = topmovies_perdecade(df)
```

```python
#printing the output
for decade in sorted(df['decade'].unique()):
    print(f"Decade: {decade}")
    print("Top Grossing Movies:")
    print(topgrossing[topgrossing['decade'] == decade][['title', 'box_office']])
    print("Top Rated Movies Based on Weighted Rating:")
    print(toprated[toprated['decade'] == decade][['title', 'weighted_rating']])
    print("\n")
```
