```python
#importing necessary modules 
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

#preprocessing

df=pd.read_csv("merged_movies2.csv")
df['starring']=df['starring'].str.strip("[]").str.replace("'","").str.split(',')
df['director']=df['director'].str.strip("[]").str.replace("'","").str.split(',')
df['profit'] = df['box_office'] - df['budget']
df['year'] = pd.to_datetime(df['release_date']).dt.year
df['decade']=(df['year']//10)*10
df=df.dropna(subset='decade')

#Weightedmeanrating calculation

def weightedmeanrating(x):
    c = df['vote_average'].mean()
    m= df['vote_count'].mean()
    v = x['vote_count']
    r = x['vote_average']
    return (v/(v+m) * r) + (m/(m+v) * c)
df['weighted_rating'] = df.apply(weightedmeanrating, axis=1)

#duplicating the rows

actordf = df.explode('starring')
directordf = df.explode('director')
```

```python
#taking the necessary attributes for calculation
actor_metrics = actordf.groupby(['decade', 'starring']).agg({'weighted_rating': 'mean', 'profit': 'sum'}).reset_index()
print(actor_metrics)
director_metrics = directordf.groupby(['decade', 'director']).agg({'weighted_rating': 'mean','profit': 'sum' }).reset_index()
```

```python
#sorting the values to get top value
topratedactor_perdecade = actor_metrics.sort_values(by=['decade', 'weighted_rating'], ascending=[True, False]).groupby('decade').head(1)
toprateddirector_perdecade = director_metrics.sort_values(by=['decade', 'weighted_rating'], ascending=[True, False]).groupby('decade').head(1)
topprofitableactor_perdecade = actor_metrics.sort_values(by=['decade', 'profit'], ascending=[True, False]).groupby('decade').head(1)
topprofitabledirector_perdecade = director_metrics.sort_values(by=['decade','profit'], ascending=[True, False]).groupby('decade').head(1)

#printing the values
print("Top-rated actors per decade:")
print(topratedactor_perdecade)
print("Top-rated directors per decade:")
print(toprateddirector_perdecade)
print("Top-profitable actors per decade:")
print(topprofitableactor_perdecade)
print("Top-profitable directors per decade:")
print(topprofitabledirector_perdecade)
```

```python
#finding top 10 values
top10rated_actors = actor_metrics.sort_values(by=['decade', 'weighted_rating'], ascending=False).groupby('decade').head(10)
top10profitable_actors = actor_metrics.sort_values(by=['decade', 'profit'], ascending=False).groupby('decade').head(10)

top10rated_directors = director_metrics.sort_values(by=['decade', 'weighted_rating'], ascending=False).groupby('decade').head(10)
top10profitable_directors = director_metrics.sort_values(by=['decade', 'profit'], ascending=False).groupby('decade').head(10)

#network function

def network(actors, directors, df, decade, network_type):
    G = nx.Graph()

#creating nodes and edges   
    for actor in actors['starring']:
        G.add_node(actor, type='actor')
    for director in directors['director']:
        G.add_node(director, type='director')
    
    for _, row in df[(df['decade'] == decade)].iterrows():
        starring = row['starring'] if isinstance(row['starring'], list) else []
        director = row['director'] if isinstance(row['director'], list) else []
        
        for actor in starring:
            for dir in director:
                if actor in G.nodes and dir in G.nodes:
                    G.add_edge(actor, dir) 
#plotting parameters
    plt.figure(figsize=(12, 8))
    pos = nx.circular_layout(G) 
    colors = ['skyblue' if G.nodes[node]['type'] == 'actor' else 'lightcoral' for node in G.nodes]
    nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='gray', node_size=3000, font_weight='bold')
    
    legend_elements = [
        Line2D([0], [0], marker='o', color='w', label='Actor', markersize=15, markerfacecolor='skyblue'),
        Line2D([0], [0], marker='o', color='w', label='Director', markersize=15, markerfacecolor='lightcoral'),
    ]
    
    plt.legend(handles=legend_elements, loc='best')
    plt.title(f"{network_type} Collaboration Network -{decade}s", fontsize=16)
    plt.axis('off')
   # plt.savefig(f'{network_type}Network{decade}s.png', bbox_inches='tight')


    plt.show()

#passing the required values to the function
for decade in sorted(df['decade'].unique()):
    actors_drated = top10rated_actors[top10rated_actors['decade'] == decade]
    directors_drated = top10rated_directors[top10rated_directors['decade'] == decade]
    actors_dgrossing = top10profitable_actors[top10profitable_actors['decade'] == decade]
    directors_dgrossing = top10profitable_directors[top10profitable_directors['decade'] == decade]

    network(actors_drated, directors_drated, df, decade, 'Top Rated')
    network(actors_dgrossing, directors_dgrossing, df, decade, 'Top Profit')

```
