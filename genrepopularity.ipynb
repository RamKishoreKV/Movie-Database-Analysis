```python
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd

#reading the file and preprocessing
df=pd.read_csv("final_movies.csv")
df['genres']=df['genres'].str.strip("[]").str.replace("'","").str.split(',')
df=df.explode('genres')
df['year']=pd.to_datetime(df['release_date']).dt.year
movies_df = df.dropna(subset=['year', 'genres'])
```

```python
genrecount_peryear = df.groupby(['year', 'genres']).size().reset_index(name='count')
#print(genrecount_peryear)
totalcounts = genrecount_peryear.groupby('genres')['count'].sum()
#print(totalcounts)
topgenres = totalcounts.sort_values(ascending=False)
topgenres = topgenres.head(10).index.tolist()
print(topgenres)
```

```python
topgenrecount_peryear = genrecount_peryear[genrecount_peryear['genres'].isin(topgenres)]
print(topgenrecount_peryear)
topgenredata = topgenrecount_peryear.pivot(index='year', columns='genres', values='count').fillna(0)
print(topgenredata)
```

```python
plt.figure(figsize=(12, 8))
for genre in topgenres:
    plt.plot(topgenredata.index, topgenredata[genre], label=genre)

plt.title('Top 10 Movie Genre Trends Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.legend(title='Genres')
plt.grid(True)
plt.tight_layout()
plt.savefig('graph.png')
```
